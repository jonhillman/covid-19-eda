## Data Acquisition

### CDC COVID-19 Deaths

#### Footnotes from data authors
Number of deaths reported in this table are the total number of deaths received and coded as of the date of analysis, and do not represent all deaths that occurred in that period. Data during this period are incomplete because of the lag in time between when the death occurred and when the death certificate is completed, submitted to NCHS and processed for reporting purposes. This delay can range from 1 week to 8 weeks or more. Percent of expected deaths is the number of deaths for all causes for this week in 2020 compared to the average number across the same week in 2017–2019. Previous analyses of 2015–2016 provisional data completeness have found that completeness is lower in the first few weeks following the date of death (<25%), and then increases over time such that data are generally at least 75% complete within 8 weeks of when the death occurred.

```python
client = Socrata("data.cdc.gov", None)
results = client.get("r8kw-7aab", limit=2000)
df_cdc = pd.DataFrame.from_records(results)

df_cdc_clean = df_cdc.copy()
df_cdc_clean.drop(['data_as_of','start_week','group','indicator', 'footnote'], axis=1, inplace=True)

df_cdc_clean['end_week'] = pd.to_datetime(df_cdc_clean['end_week']).dt.date
df_cdc_clean['state'] = df_cdc_clean['state'].astype(str)
df_cdc_clean['covid_deaths'] = df_cdc_clean['covid_deaths'].astype(np.float32)
df_cdc_clean['total_deaths'] = df_cdc_clean['total_deaths'].astype(np.float32)
df_cdc_clean['percent_of_expected_deaths'] = df_cdc_clean['percent_of_expected_deaths'].astype(np.float16)
df_cdc_clean['pneumonia_deaths'] = df_cdc_clean['pneumonia_deaths'].astype(np.float16)
df_cdc_clean['pneumonia_and_covid_deaths'] = df_cdc_clean['pneumonia_and_covid_deaths'].astype(np.float16)
df_cdc_clean['influenza_deaths'] = df_cdc_clean['influenza_deaths'].astype(np.float16)
df_cdc_clean['pneumonia_influenza_or_covid_19_deaths'] = df_cdc_clean['pneumonia_influenza_or_covid_19_deaths'].astype(np.float16)

df_cdc_clean = df_cdc_clean.fillna(0)
```

```python
us_pop_by_age = pd.read_csv('data/US_pop_by_single_age.csv')
us_pop_by_age.drop(np.arange(103,len(us_pop_by_age),1), inplace=True)
us_pop_by_age = us_pop_by_age.groupby(by='AGE').sum()
us_pop_by_age.drop(999, inplace=True)
us_pop_by_age = us_pop_by_age[['POPESTIMATE2019']]
us_pop_by_age.rename(columns={'POPESTIMATE2019':'population'}, inplace=True)
us_pop_by_age = us_pop_by_age.reset_index()
indexes = []
total = 0
for index, row in us_pop_by_age.iterrows():
    if row['AGE'] > 85:
        indexes.append(index)
        total += row['AGE']
us_pop_by_age.drop(indexes, inplace=True)
us_pop_by_age.iloc[85]['population'] += total
us_pop_by_age.rename(columns={'AGE':'age'}, inplace=True)
us_pop_by_age.head()
```

### US CDC Deaths by Age

```python
results = client.get("3apk-4u4f", limit=180)

df_cdc_ages = pd.DataFrame.from_records(results)

df_cdc_ages = df_cdc_ages[['sex', 'age_years', 'total_deaths', 'covid19_deaths']]
for index, row in df_cdc_ages.iterrows():
    df_cdc_ages.loc[index, 'age_years'] = int(str(row['age_years']).replace(' Years','').replace('<1 Year','0').replace(' and over',''))
df_cdc_ages['total_deaths'] = df_cdc_ages['total_deaths'].astype(np.int64)
df_cdc_ages['covid19_deaths'] = df_cdc_ages['covid19_deaths'].astype(np.int64)
df_cdc_ages['covid_percent_of_total'] = df_cdc_ages['covid19_deaths']/df_cdc_ages['total_deaths']

covid_deaths = df_cdc_ages['covid19_deaths'].sum()
total_deaths = df_cdc_ages['total_deaths'].sum()
df_cdc_ages_sex = df_cdc_ages.groupby(by='sex')
df_cdc_ages_male = df_cdc_ages_sex.get_group('Male').copy()
df_cdc_ages_female = df_cdc_ages_sex.get_group('Female').copy()

df_cdc_ages_male_covid_total = df_cdc_ages_male['covid19_deaths'].sum()
df_cdc_ages_male_total = df_cdc_ages_male['total_deaths'].sum()
df_cdc_ages_male['covid_percent_of_total_by_age'] = df_cdc_ages_male['covid19_deaths']/df_cdc_ages_male['total_deaths']
df_cdc_ages_male['covid_percent_of_males_by_age'] = df_cdc_ages_male['covid19_deaths']/df_cdc_ages_male_covid_total
df_cdc_ages_male['total_percent_of_males_by_age'] = df_cdc_ages_male['total_deaths']/df_cdc_ages_male_total

df_cdc_ages_female_covid_total = df_cdc_ages_female['covid19_deaths'].sum()
df_cdc_ages_female_total = df_cdc_ages_female['total_deaths'].sum()
df_cdc_ages_female['covid_percent_of_total_by_age'] = df_cdc_ages_female['covid19_deaths']/df_cdc_ages_female['total_deaths']
df_cdc_ages_female['covid_percent_of_females_by_age'] = df_cdc_ages_female['covid19_deaths']/df_cdc_ages_female_covid_total
df_cdc_ages_female['total_percent_of_females_by_age'] = df_cdc_ages_female['total_deaths']/df_cdc_ages_female_total
```

## Feature Engineering

```python
df_cdc_eng = df_cdc_clean.copy()
```

### deaths_week_count
* This is the number of weeks since the first death was reported per state.
* This feature can be used as the basis to compare the progression of deaths from state-to-state over time.

```python
df_cdc_eng['deaths_week_count'] = 0
state = ''
week_count = 0
deaths = 0
for index, row in df_cdc_eng.iterrows():
    if row['state'] != state:
        state = row['state']
        week_count = 0
        deaths = 0
    if row['covid_deaths'] > 0 or week_count > 0:
        week_count += 1
        deaths += row['covid_deaths']
    df_cdc_eng.loc[index, 'deaths_week_count'] = week_count
```

### mask_mandate_severity
* A number 0-3 (less to more severe) representing the severity of lockdown policy by state
* 0 until mandate took effect, single number thereafter

```python
state_mandates = {}
for index, row in df_mask_mandates.iterrows():
    state_mandates[row['state']] = [row['mask_mandate_severity'], row['mask_mandate_date']]
df_cdc_eng['mask_mandate_severity'] = 0
df_cdc_eng['mask_mandate'] = 0
for index, row in df_cdc_eng.iterrows():
    if row['state'] not in state_mandates:
        continue
    if row['end_week'] >= state_mandates[row['state']][1]:
        df_cdc_eng.loc[index, 'mask_mandate_severity'] = state_mandates[row['state']][0]
        df_cdc_eng.loc[index, 'mask_mandate'] = 1
    else:
        df_cdc_eng.loc[index, 'mask_mandate_severity'] = 0
        df_cdc_eng.loc[index, 'mask_mandate'] = 0
df_cdc_eng.head()
```

### covid_deaths_per_1m
* covid deaths per million in the population per day

```python
df_cdc_eng['covid_deaths_per_1m'] = 0
for index, row in df_cdc_eng.iterrows():
    state = row['state']
    if state.lower() not in df_mobility.index.values:
        continue
    df_cdc_eng.loc[index, 'covid_deaths_per_1m'] = row['covid_deaths']/df_us_pop.loc[state]['population']*1000000
df_cdc_eng.head()
```

### mobility
* sourced from Apple Mobility data, weekly mean values from 2020/2/1

```python
df_cdc_eng['mobility'] = 0
for index, row in df_cdc_eng.iterrows():
    state = row['state']
    if state.lower() not in df_mobility.index.values:
        continue
    end_week = row['end_week']
    mobility = df_mobility.loc[state.lower(), str(end_week)]
    df_cdc_eng.loc[index,'mobility'] = mobility
df_cdc_eng.head()
```

### teacher_student_population_ages
* teacher and student populations estimated from NECS data and Census population data

```python
total_teachers = 3385200
#ages
#25-30 = 15.6%
#30-49 = 54.0%
#50-54 11.9%
#55-67 = 18.8%
teachers_age_distribution = {
    30:((total_teachers*.156),5),
    49:((total_teachers*.54),20),
    54:((total_teachers*.119),5),
    55:((total_teachers*.185),12)
}
keys = list(teachers_age_distribution.keys())
teachers_by_age = {}
key = 0
for i in np.arange(25, 67, 1):
    if i > keys[key] and i < 55:
        key+=1
    pop, bins = teachers_age_distribution[min(max(i,keys[key]),55)]
    teachers_by_age[i] = round(pop/bins)

total_students = 56000000
students_by_age = {}
total_student_pop = us_pop_by_age.iloc[np.arange(5,19)]['population'].sum()
for i in np.arange(5,19,1):
    students_by_age[i] = int(total_students*(us_pop_by_age.iloc[i]['population']/total_student_pop))
```

### State Totals

```python
df_cdc_eng_totals = df_cdc_eng.groupby(by=['state']).agg('sum')
df_cdc_eng_totals.reset_index(inplace=True)
df_cdc_eng_totals
```

#### covid_deaths_per_100k

```python
df_cdc_eng_totals['covid_deaths_per_1m'] = 0
df_cdc_eng_totals['total_deaths_per_1m'] = 0
for index, row in df_cdc_eng_totals.iterrows():
    df_cdc_eng_totals.loc[index, 'covid_deaths_per_1m'] = row['covid_deaths']/df_us_pop.loc[row['state']]['population']*1000000
    df_cdc_eng_totals.loc[index, 'total_deaths_per_1m'] = row['total_deaths']/df_us_pop.loc[row['state']]['population']*1000000
df_cdc_eng_totals.head()
```

#### Create separate dataframes for each state

```python
states_by_name = {}
state_death_totals = []
for name, group in df_cdc_eng.groupby('state'):
    if name != 'United States':
        total = group['covid_deaths'].sum()
        masks = group['mask_mandate_severity'].sum()
        deaths_per_1m = total/df_us_pop.loc[name]['population']*1000000
        state_death_totals.append((name, total, masks, deaths_per_1m))
        states_by_name[name] = group
state_death_totals.sort(key=lambda tup: tup[3], reverse=True)
```

```python
us_covid_deaths = 0
us_total_deaths = 0
us_total_pop = df_us_pop['population'].sum()
for name, group in df_cdc_eng.groupby('state'):
    if name == 'United States':
        us_covid_deaths = group['covid_deaths'].sum()
        us_total_deaths = group['total_deaths'].sum()
columns = ['state', 'total_covid_deaths', 'percent_of_covid_deaths', 'percent_of_all_deaths', 'covid_deaths_per_1m','total_deaths_per_1m','percent_of_population', 'population_density', 'mask_mandate_severity', 'mobility'] 
data = [[]]
for name, group in df_cdc_eng.groupby('state'):
    if not group.empty and name != 'United States':
        covid_total = group['covid_deaths'].sum()
        total = group['total_deaths'].sum()
        mask_mandate_severity = group['mask_mandate_severity'].sum()
        population = df_us_pop.loc[group['state']]['population'].values[0]
        covid_deaths_per_1m = covid_total/population * 1000000
        total_deaths_per_1m = total/population * 1000000
        density = df_us_pop.loc[group['state']]['density_per_sq_mile'].values[0]
        mobility = 100.0
        if name.lower() in df_mobility.index.values:
            mobility = df_mobility.loc[name.lower()].values.mean()
        data.append([name, covid_total, covid_total/us_covid_deaths, total/us_total_deaths, covid_deaths_per_1m, total_deaths_per_1m ,population/us_total_pop, density, mask_mandate_severity, mobility])
df_cdc_cum = pd.DataFrame(data, columns=columns)
df_cdc_cum.dropna(inplace=True)
df_cdc_cum.sort_values(by=['covid_deaths_per_1m'], inplace=True, ascending=False)
df_cdc_cum.index = np.arange(1, len(df_cdc_cum)+1)
df_cdc_cum.head()
```

```python
#focused protection as raised in the Great Barrington Declaration
focused_protection_age_start = 65
focused_protection_factors = list(np.ones(focused_protection_age_start))
focused_protection_reduction_max = .9 #% exposure reduction at age 85, linearly increases from focused_protection_age_start
focused_protection_step = focused_protection_reduction_max/(85-focused_protection_age_start+1)
avg_reduction = np.arange(1, .1, -focused_protection_step).mean()
for i in np.arange(1, .1, -focused_protection_step):
    focused_protection_factors.append(i)

traditional_herd_threshold = .7
death_predictions_by_age = us_pop_by_age.copy()

#https://www.medrxiv.org/content/10.1101/2020.07.23.20160895v5
# log(IFR) = -7.53 + 0.119 * age
#IFR = exp(-7.53 + 0.119 * age) # this yields 1.3% mean IFR, however WHO has this figure at .13% (1/10th)

for index, row in death_predictions_by_age.iterrows():
    ifr = np.exp(-7.53 + 0.119 * row['age'])/1000
    death_predictions_by_age.loc[index,'ifr'] = ifr

ifr_mean = death_predictions_by_age['ifr'].mean()

#% of US population already infected, removed from total population
#this is inferred as:
#infected = existing fatalities/WHO mean IFR * US population
existing_covid_deaths = df_us_totals_by_week['covid_deaths'].sum()
us_pop = us_pop_by_age['population'].sum()
exposed_population_percent = df_us_totals_by_week['covid_deaths'].sum()/(us_pop*ifr_mean)

death_predictions_by_age['deaths_max'] = death_predictions_by_age['population'] * (1-exposed_population_percent) * death_predictions_by_age['ifr']
death_predictions_by_age['deaths_focused_protection'] = death_predictions_by_age['population'] * (1-exposed_population_percent) * traditional_herd_threshold * focused_protection_factors[index] * death_predictions_by_age['ifr']

upper_threshold = int(death_predictions_by_age['deaths_max'].sum())
focused_protection = death_predictions_by_age['deaths_focused_protection'].sum()

us_weekly_deaths_means = []
interval = -2
end = -1
for i in range(1,16):
    us_weekly_deaths_means.append(df_us_totals_by_week['covid_deaths'][end+interval:end].mean())
    end += interval
us_weekly_deaths_means = us_weekly_deaths_means[::-1]
weekly_mean_changes = []
for i in range(1, len(us_weekly_deaths_means)-1):
    weekly_mean_changes.append((us_weekly_deaths_means[i] - us_weekly_deaths_means[i-1])/us_weekly_deaths_means[i-1])

change_per_week = weekly_mean_changes[len(weekly_mean_changes)-1]
weekly_deaths_predicted = [int(us_weekly_deaths_means[len(us_weekly_deaths_means)-1])]
today = dt.datetime.today().date()
end_of_year = pd.to_datetime('2020-12-31').date()
weeks_remaining = (end_of_year - today).days // 7
for i in range(1, weeks_remaining):
    weekly_deaths_predicted.append(int((1+change_per_week) * weekly_deaths_predicted[i-1]))

#statement
print(f'With a mean infection fatality rate of {"{:.3%}".format(ifr_mean)} (as indicated by the World Health Organization) \
and exposure reduction for ages {focused_protection_age_start} and up in line with what is currently practiced, \
if the amount of the population already infected is {"{:.1%}".format(exposed_population_percent)} \
(as inferred from US Population of {"{:n}".format(int(us_pop))}, mean IFR as above, and CDC Provisional Death Counts), \
we can expect to see a total of {"{:n}".format(int(existing_covid_deaths+focused_protection))} deaths with COVID-19 in the US in 2020. \
This means we can expect to see another {"{:n}".format(int(focused_protection))} COVID-19 deaths in \
the remaining {weeks_remaining} weeks of 2020, \
with the following deaths per week: {weekly_deaths_predicted}.')                             
                                                                       
                                         
```
